## What is file Handling??

File Handling is used to read and write data in files.

Python can:

Create files
-Read files-
-write files
-Add new data to files 

```python
file=open("demo.txt","w")

file.write("Hello World")

file.close()

print('file created successfully ')
```

# Reading a File

we can read()to read file content.

```python
file = open("demo.txt","r")

print(file.read())

file.close()
```

# Writing New Content 

Write mode  replace old content 

```python
file= open("demo.txt","w")

file.write (" I am learning Python")

file.close()

print("Data Written")
```

```python
file = open ("demo.txt","r")

print (file.read())

file.close()
```

# Append Mode 

Append mode adds new content to an existing file.

Mode=a

```python
file= open("demo.text","a")

file.write("\nWelcome to file Handling")

file.close()

print("New Data Added")
```

```python
file=open("demo.txt","r")

print(file.read())

file.close()
```

# readline() 

reading () reads one line at a time .

```python
file = open("demo.txt","r")

print(file.readline())

file.close()
```

# with open()

with open()automatically closes the file.

This is the recommended way to work with files .

```python
with open("demo.txt","r") as file:
    print(file.read())
```

# Writing Using with open()

```python
with open ("newfile.txt","w")as file:
    file.write("Learning Python file Handling")

print("File Saved")
```

```python
with open("newfile.txt","r")as file:
    print(file.read())
```
