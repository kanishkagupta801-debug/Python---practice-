## What is Functions?

A function is a reusable block of code.

functioning help us avoid writting the same code again and again .

*:Benefits:*
#code reusability
#Better Organiztion
#easy Debugging
#less repetition 

```python
def greet():
    print("Hello, World!")
greet()
```

###EXPLAINATION

-def is used to create a function.
-greet is the function name
-greet() is used to call the function.

```python
# FUNCTION CALLED MULTIPLE TIMES
def greet():
    print("Hello, World!")

greet()
greet()
greet()
greet() 
```

# FUNCTION WITH PARAMETER

Parameter allow us to pass values into a function 

```python
def greet(name):        
    print("Hello", name)

greet("kanishka")
greet("radha")
```

###PARAMETER

"name" is a parameter.

The value passed during functions call is called an argument .

```python
def city(city_name):
    print("I live in", city_name)

city("New York")
city("Los Angeles")
```

#RETURN STATEMENT 

The return statement is used to send a value back from a function.

unlike print().return gives us a value that can be stored and reuse 

```python
def square(num):
    return num * num

result = square(9)

print (result)
```

### Diffrence Between Print and /return 

print() only diplays output 

 return sends output back stores it for future uses 

```python
def add (a,b):
    return a +b

answer = add (30,59)

print (answer)
```
