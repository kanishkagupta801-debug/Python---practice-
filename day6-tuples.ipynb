# WHAT IS TUPLES???

A tuples is a collection of items.

Tuples aare ordered and immutable .

Immutable means the values cannot thee change after creation.


```python
fruits=("Apple","Mango","Banana")
print(fruits)
```

Tuples are similar to lists but they cannot be changed.
We use () insteadd of [].

```python
fruits=["Apple","Mango","Banana" ]

print(fruits[0])
print(fruits[1])
print(fruits[2])
```

## Tuples Indexing

Indexing starts from 0 just like lists.

```python
fruits=["Apple","Mango","Banana" ]
print(len(fruits))
```

```python
fruits=("Apple","Mango","Banana")

for fruit in  fruits:
    print(fruit)
```

```python
numbers=[ 1,2,3,4,5 ]
for num in numbers:
    print(num)
```

# Tuples - Important topic

Tuples are used when data should not change .

Example:
-Days of week
-Months
-Fixed Data like country codes

```python
days=("Monday","Tuedsay","Wednesday","Thrusday","Friday","Saturday","Sunday")
for day in days:

    print(day)
```
