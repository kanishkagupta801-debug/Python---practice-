# Acessing Dictionary values

we can acess values using keys.

Syntax:

dictionary_name["key"]

```python
student={
    "name":"Kanishka",
    "age":20,
    "course":"python"

}

print(student["name"])
print(student["age"])
```

keys are used to acess values.

"name" returns kanishka.
"age" returns 20.



# Adding New Items :

We can add new- mey value pairs to a dictionary .

```python
student={
    "name":"Kanishka", 
    "age":20,
}

student["city"]="Lucknow"

print(student)
```

# Updating Values :

Dictionary values can be updated using keys.

```python
student={
    "name":"Kanishka",
    "age":20,
}

student["age"]=21  

print(student)
```

# Loop Through a Dictionary:

 We can use a for loop to a acess  dictionary keys one by one .

```python
student={
    "name":"Kanishka",
    "age":20,
    "course":"python"  
}
for key in student :
    print(key)
```

# Acessing Keys :

keys() returns all keys from a dictionary .

```python
student={
    "name":"Kanishka",
    "age":20,
}       

print(student.keys())
```

# ACESSING KEYS AND VALUES TOGETHER

items() returns both keys and values 

```python
student={
    "name":"Kanishka",
    "age":20,
    "course":"python"
}
for key in student.items():
    print(key)
```

# Removing items from dictionary 
 we can remove items using pop().
 

```python
student={
    "name":"Kanishka",
    "age":20,
}
student.pop("age")

print(student)  
```
