# Loops in Python

Loops are used in to repeat a block code multiple times 

  # First Loops Program 

   This Program pronts number from 0 to 4 

```python
for i in range (5):
    print(i)
```

# Name printing Multiple Lines

This program prints the name multiples times.

```python
for i in range(5):
    print("kanishka")
```

# Emoji Program 

This Program Prints Emojis using loops 

```python
for i in range (5):
    print ("😊")
```

# Countdown program
This program prints numbers from 5 to 1.

```python
for i in range(5,0,-1):
    print(i)
```

# Pattern Program 
This Program Prints a heart Pattern.

```python
for i in range (1, 6):
    print("❤️"*i)
```
