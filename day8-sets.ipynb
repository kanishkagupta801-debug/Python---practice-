## What is a sets?

A set is a collection of unique items.

Sets:
-Do not Allow duplicate values
-Are unordered
-Are writing using()

```python
fruits=("Apple","Mango","Banana")

print(fruits)
```

# Duplicate values

```python
numbers= (1,2,3,3,4,4,5)

print(numbers)
```

Sets keep only unique values.
Duplicate items are removed automatically

```python
fruits=("apple","mango","banana")
print(len(fruits))
```

# Loop Through a Set 

We can use a for loop to acess set items .

```python
fruits=("apple","mango","banana")

for fruit in fruits:
    print(fruit)
```

```python
cities=("Lucknow","Delhi","Mumbai","Delhi")
print(cities)
```

# Adding Items to a Set :

We can add () to add a new item toa set .

```python
fruits = {"Apple","Mango"}

fruits.add("Banana")

print(fruits)
```

# Removing items from a set

We use remove () to delete an item from a set

```python
Names={"John","Peter","David"}

Names.remove("Peter")

print(Names)
```

# Union of sets:

Union combines all unique value from two sets.

```python
set1 ={1,2,3}
set2 ={3,4,5}

print(set1.union(set2))
```

# Intersection of sets :

Intersection returns common values from both set.

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

print(set1.intersection(set2))
```

# Discard()Method 

discard()removes an item from a set

Unlike remove(),discard()does not give an error if item is not present .

```python
A = {1, 2, 3}
A.discard(2)

print(A)
```
