# STRING LENGTH

We can use len() to count characters in a string.

```python
name="Kanishka"

print(len(name))
```

# Acess Characters

Each character has an index position.

Index starts from 0.

```python
name="Kanishka"

print(name[0])
print(name[1])
print(name[2])
print(name[3])
print(name[4])
print(name[5])
print(name[6])
```

# Negative Indexing

Negative Indexing start from the end .

```python
name="Kanishka"

print(name[-1])
print(name[-2])
print(name[-3])
```

# Start Slicing

slicing extractas a part  of a strings.

```python
text="Python"

print(text[0:2])
print(text[2:5])
```

# UPPER CASE

upper() converts all character to uppercase.

```python
text="Hello, World!"

print(text.upper())
print(text.lower())
```

# REPLACE TEXT

```python
text=" I love Python"

print(text.replace("Python","Coding"))
```

# count() Method:

count() counts how many times a character or word appears in a string .

```python
text = "Python is great. Python is easy to learn. I love Python."

print(text.count("Python"))
```

# find() Method :

find() returns the posotion of a word or character .

```python
text="i love python"

print(text.find("Python"))
```

# String Concatenation

we can join strings using =.

```python
first_name="Kanishka"
last_name="Shrestha"

full_name=first_name+""+last_name

print(full_name)
```
