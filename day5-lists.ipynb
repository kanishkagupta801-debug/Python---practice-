# Acessing  Lists items 

Each item in a list has an index.
 Index starts from 0.
 

```python
fruits=["Apple" ,"Mango"," Banana"]

print (fruits[0])   
print (fruits[1])
print (fruits[2])       
```

```python
fruits= ["apple","mango","banana" ]

print(len(fruits))
```

# ADDING ITEMS TO  A LIST 

Python allows us to add new items to alist using append[].

```python
fruits=["Apple","Mango","Banana" ]
fruits.append("Grapes")
print(fruits)
```

###append() add a new item at the end of the list.

# INSERTING ITEMS

insert() is used to add an item on a specific position .

```python
colors=["Red","Green","Blue"]
colors.insert(1,"Yellow")

print(colors)
```

# REMOVING ITEMS 

remove() removes on item by value 

```python
fruits=["Apple","Mango","Banana" ]
fruits.remove("Mango")  

print(fruits)
```

# POP METHOD 

pop() removes an item using its index 

```python
fruits=["Apple","Mango","Banana"]

fruits.pop(2)

print(fruits)
```

# LOOP THROUGH A LIST 

A loop can be used to acess each item in a list one by one.
The for loop is commonly used with lists.

```python
fruits=["Apple","Mango","Banana" ]

for fruit in fruits:
    print (fruit)   
    
```

```python
students=["Kanishka","Anshul","Shreya"]
for student in students:
    print("Hello", student)
```

# LIST LENTGH WITH LOOP 

We  can use len() with range () to acess list index .

```python
fruits=["Apple","Mango","Banana" ]

for i in range(len(fruits)):
    print(i,fruits[i])
```

```python
marks=[ 80,90,76,54,65 ]
total = 0 

for mark in marks:
    total = total + mark

print ("Total marks:", total)
```

```python
numbers =[ 10,20,30,40,50]

for num in numbers:
    print(num*2)
```

```python
cities=["Delhi","Mumbai","Bangalore"]

for city in cities:
    print("Welcome to", city)   
```
