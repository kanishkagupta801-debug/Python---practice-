#  Specific Exception Handling

Python allows us to handle specific errors seprately.

Common Error :

-ZeroDivisionError
-ValueError
-NameError

```python
try:
    num=10/10

except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
try:
   age=int("hello")

except ValueError:
   print("Pleaase enter numbers only")
```

```python

try:
    num=10/2

except:
    print("Error")

else:
    print("No Error")
```

```python
try:
    num=10/0

except:
    print("Error")

finally:
    print("Program finished")
```

## 🎯 MNI PROJECT-SAFE CALCULATOR

```python
try:
   num1 =10
   num2 =0

   result=num1/num2

except ZeroDivisionError:
   print("Cannot divide by zero")

else:
   print("Result=",result)

finally:
   print("Calculator Closed")
```
