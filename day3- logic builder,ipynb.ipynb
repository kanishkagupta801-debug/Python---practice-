# While Loops

A while loops repeat code while a condition is true .

```python
i=1
while i<=5:
    print(i)
    i=i+1
```

# Nested Loop 

A nested loop means a loop inside another loop.
the outer loop controls thr number of repitions,while the inner loop runs complety for each of the iteration of the outer loop  Nested loop are commonly used in pattern programs tables and matrix operations .

```python
for i in range (1,4):
    for j in range (1,4):
        print(i,j)
```

#Pattern Program 

Nested loops can be used to create Patterns

The inner loop Prints stars on the same line .
The outer loop  controls the number of rows

```python
for i in range (3):
    for j in range (5):
        print("*",end="")
    print()
```

```python
#Code CELL

for i in range(4):
    for j in range(4):
        print("#",end="")
    print()
```

#Break Statement to


The break statement is used to stop a loop immediately.

when break is executed,the loop terminates and the remaining iteration are skipped.

```python
for i in range (1,11):
    if i == 9:
        break
    print(i)
```

# Continue Statement

The continue statement skip the current iteration and moves to the next iteration of the loop.

```python
for i in range (1,9):
    if i == 6:
        continue
    print(i)
```
